Due to the missing values of City represent less than 4% of the total data, we will drop the rows with missing values in the City column.

```python
df = df.dropna(subset=['City'])
df
```

```python
df.isna().sum()
```

The missing values of Column name still in the dataset due to ......

```python
df_final = df.copy()
df_final['Customer Name'] = df_final['Customer Name'].fillna('Unknown')
```

```python
df_final.isna().sum()
```

```python
df2 = df_final.copy()

# Convert Dates
df2['Order Date'] = pd.to_datetime(df2['Order Date'], format= 'mixed')
df2['Ship Date'] = pd.to_datetime(df2['Ship Date'], format= 'mixed')


# Convert Financials (Cleaning strings like '$1,200.00' first)
df2['Sales'] = pd.to_numeric(df2['Sales'].replace(r'[\$,]', '', regex=True), errors='coerce')
df2['Profit'] = pd.to_numeric(df2['Profit'].replace(r'[\$,]', '', regex=True), errors='coerce')

# Convert Quantity to Integer
df2['Quantity'] = pd.to_numeric(df2['Quantity'], errors='coerce').fillna(0).astype(int)

# Clean up Postal Code (removing decimal .0 and treating as string)
df2['Postal Code'] = df2['Postal Code'].fillna(0).astype(int).astype(str)

# Optimize Categorical columns
categorical_cols = ['Ship Mode', 'Segment', 'Region', 'Category', 'Sub-Category']
df2[categorical_cols] = df2[categorical_cols].astype('category')

# Verify the changes
print(df2.info())
```

```python

# Check for any dates that couldn't be parsed at all
print(f"Missing Order Dates: {df2['Order Date'].isna().sum()}")
print(f"Missing Ship Dates: {df2['Ship Date'].isna().sum()}")
```

```python
##Story 1
required_columns = ["Region", "Category", "Sales", "Profit"]

missing_columns = [col for col in required_columns if col not in df2.columns]

if missing_columns:
    print("Missing columns:", missing_columns)
else:
    print("All required columns are present.")
```

```python
##Story 1
sales_by_region = df2.groupby("Region")["Sales"].sum().sort_values(ascending=False)

sales_by_region
```

```python
##Story 1
profit_by_region = df2.groupby("Region")["Profit"].sum().sort_values(ascending=False)

profit_by_region
```

```python
sales_by_category = df2.groupby("Category")["Sales"].sum().sort_values(ascending=False)

sales_by_category
```

```python
profit_by_category = df2.groupby("Category")["Profit"].sum().sort_values(ascending=False)

profit_by_category
```

```python
#Story 1
summary = {
    "sales_by_region": sales_by_region,
    "profit_by_region": profit_by_region,
    "sales_by_category": sales_by_category,
    "profit_by_category": profit_by_category
}

summary
```
