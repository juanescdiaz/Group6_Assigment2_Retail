```python
df_final = df.copy()
df_final['Customer Name'] = df_final['Customer Name'].fillna('Unknown')
```

```python
df_final.isna().sum()
```

```python
df2 = df_final.copy()

# Convert Dates
df2['Order Date'] = pd.to_datetime(df2['Order Date'], format= 'mixed')
df2['Ship Date'] = pd.to_datetime(df2['Ship Date'], format= 'mixed')


# Convert Financials (Cleaning strings like '$1,200.00' first)
df2['Sales'] = pd.to_numeric(df2['Sales'].replace(r'[\$,]', '', regex=True), errors='coerce')
df2['Profit'] = pd.to_numeric(df2['Profit'].replace(r'[\$,]', '', regex=True), errors='coerce')

# Convert Quantity to Integer
df2['Quantity'] = pd.to_numeric(df2['Quantity'], errors='coerce').fillna(0).astype(int)

# Clean up Postal Code (removing decimal .0 and treating as string)
df2['Postal Code'] = df2['Postal Code'].fillna(0).astype(int).astype(str)

# Optimize Categorical columns
categorical_cols = ['Ship Mode', 'Segment', 'Region', 'Category', 'Sub-Category']
df2[categorical_cols] = df2[categorical_cols].astype('category')

# Verify the changes
print(df2.info())
```

```python
# Check for any dates that couldn't be parsed at all
print(f"Missing Order Dates: {df2['Order Date'].isna().sum()}")
print(f"Missing Ship Dates: {df2['Ship Date'].isna().sum()}")
```

### Segment Based Analysis 

```python
def compare_segments(df, seg_a, seg_b, start=None, end=None):

    # Define the segment mask
    mask = df['Segment'].isin([seg_a, seg_b])
    
    # Add optional Date filtering
    if start or end:
        start_dt = pd.to_datetime(start) if start else df['Order Date'].min()
        end_dt = pd.to_datetime(end) if end else df['Order Date'].max()
        
        lower = min(start_dt, end_dt)
        upper = max(start_dt, end_dt)
        
        mask = mask & (df['Order Date'] >= lower) & (df['Order Date'] <= upper)

        print( ' Analysis Period: ' + lower.strftime('%Y-%m-%d') + ' to ' + upper.strftime('%Y-%m-%d') + ' ')   
    else:    
        print(' Analysis Period: Full Dataset  ')
    
    df_filtered = df.loc[mask]
        

    # Aggregate all four required metrics
    comparison = df_filtered.groupby('Segment', observed=True).agg(
        Avg_Sales=('Sales', 'mean'),
        Total_Sales=('Sales', 'sum'),
        Avg_Profit=('Profit', 'mean'),
        Total_Profit=('Profit', 'sum')
    )

    # Calculate the Difference
    if len(comparison) >= 2:
        delta = comparison.loc[seg_a] - comparison.loc[seg_b]
        comparison.loc[f'Delta ({seg_a} - {seg_b})'] = delta
    
    return comparison

# Ensure professional financial formatting
pd.options.display.float_format = '{:,.2f}'.format
```

```python
# TEST 1: Full Dataset (No dates provided)
report_all = compare_segments(df2, 'Corporate', 'Consumer')
print(report_all)
```

```python
# TEST 2: Specific range (Only 2017)
report_2017 = compare_segments(df2, 'Corporate', 'Consumer', start='2017-01-01', end='2017-12-31')
print(report_2017)
```
