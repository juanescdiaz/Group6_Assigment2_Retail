```python
df_final = df.copy()
df_final['Customer Name'] = df_final['Customer Name'].fillna('Unknown')
```

```python
df2 = df_final.copy()

# Convert Dates
df2['Order Date'] = pd.to_datetime(df2['Order Date'], format= 'mixed')
df2['Ship Date'] = pd.to_datetime(df2['Ship Date'], format= 'mixed')


# Convert Financials (Cleaning strings like '$1,200.00' first)
df2['Sales'] = pd.to_numeric(df2['Sales'].replace(r'[\$,]', '', regex=True), errors='coerce')
df2['Profit'] = pd.to_numeric(df2['Profit'].replace(r'[\$,]', '', regex=True), errors='coerce')

# Convert Quantity to Integer
df2['Quantity'] = pd.to_numeric(df2['Quantity'], errors='coerce').fillna(0).astype(int)

# Clean up Postal Code (removing decimal .0 and treating as string)
df2['Postal Code'] = df2['Postal Code'].fillna(0).astype(int).astype(str)

# Optimize Categorical columns
categorical_cols = ['Ship Mode', 'Segment', 'Region', 'Category', 'Sub-Category']
df2[categorical_cols] = df2[categorical_cols].astype('category')

# Verify the changes
print(df2.info())
```

```python
# Check for any dates that couldn't be parsed at all
print(f"Missing Order Dates: {df2['Order Date'].isna().sum()}")
print(f"Missing Ship Dates: {df2['Ship Date'].isna().sum()}")
```

```python
pd.options.display.float_format
def compare_segments(df, seg_a, seg_b):
   
    # Create a deep copy so the original df is untouched
    temp_df = df.copy()
    
    
    temp_df['Sales'] = pd.to_numeric(temp_df['Sales'].replace(r'[\$,]', '', regex=True), errors='coerce').fillna(0)
    temp_df['Profit'] = pd.to_numeric(temp_df['Profit'].replace(r'[\$,]', '', regex=True), errors='coerce').fillna(0)
    temp_df['Order Date'] = pd.to_datetime(temp_df['Order Date'], format='mixed')

    
    comparison = temp_df[temp_df['Segment'].isin([seg_a, seg_b])].groupby('Segment').agg(
        Avg_Profit=('Profit', 'mean'),
        Total_Sales=('Sales', 'sum')
    )
    
    
    delta = comparison.loc[seg_a] - comparison.loc[seg_b]
    comparison.loc[f'Difference ({seg_a} - {seg_b})'] = delta
    
    return comparison
```

```python
pd.options.display.float_format = '{:,.2f}'.format

report = compare_segments(df2, 'Corporate', 'Consumer')


print("--- Strategy Consultant: Segment Comparison Report ---")
print(report)
```

```python
import pandas as pd

def compare_segments_advanced(df, seg_a, seg_b, start=None, end=None):
    # 1. Use the requested default dates if none are provided
    # Default start: Jan 1st, 2024 | Default end: June 1st, 2018
    start_dt = pd.to_datetime(start) if start else pd.to_datetime('2024-01-01')
    end_dt = pd.to_datetime(end) if end else pd.to_datetime('2018-06-01')
    
    # 2. Chronological Check: Ensure the smaller date is always the 'start' for the filter
    # This handles the fact that Jan 2024 is after June 2018
    lower_bound = min(start_dt, end_dt)
    upper_bound = max(start_dt, end_dt)

    # 3. Filter using 'Order Date' and the selected Segments
    mask = (
        (df['Segment'].isin([seg_a, seg_b])) & 
        (df['Order Date'] >= lower_bound) & 
        (df['Order Date'] <= upper_bound)
    )
    df_filtered = df.loc[mask]

    # 4. Aggregate Metrics
    comparison = df_filtered.groupby('Segment', observed=False).agg(
        Avg_Profit=('Profit', 'mean'),
        Total_Sales=('Sales', 'sum')
    )

    # 5. Calculate Difference (Delta)
    if len(comparison) >= 2:
        delta = comparison.loc[seg_a] - comparison.loc[seg_b]
        comparison.loc[f'Delta ({seg_a} - {seg_b})'] = delta
    
    return comparison

# Apply the global format fix so the 'Total_Sales' looks like money
pd.options.display.float_format = '{:,.2f}'.format

# Try it out on your df_final
report = compare_segments_advanced(df2, 'Corporate', 'Consumer')
print(report)
```
