```python
df_final = df.copy()
df_final['Customer Name'] = df_final['Customer Name'].fillna('Unknown')
```

```python
df_final.isna().sum()
```

```python
df2 = df_final.copy()

# Convert Dates
df2['Order Date'] = pd.to_datetime(df2['Order Date'], format= 'mixed')
df2['Ship Date'] = pd.to_datetime(df2['Ship Date'], format= 'mixed')


# Convert Financials (Cleaning strings like '$1,200.00' first)
df2['Sales'] = pd.to_numeric(df2['Sales'].replace(r'[\$,]', '', regex=True), errors='coerce')
df2['Profit'] = pd.to_numeric(df2['Profit'].replace(r'[\$,]', '', regex=True), errors='coerce')

# Convert Quantity to Integer
df2['Quantity'] = pd.to_numeric(df2['Quantity'], errors='coerce').fillna(0).astype(int)

# Clean up Postal Code (removing decimal .0 and treating as string)
df2['Postal Code'] = df2['Postal Code'].fillna(0).astype(int).astype(str)

# Optimize Categorical columns
categorical_cols = ['Ship Mode', 'Segment', 'Region', 'Category', 'Sub-Category']
df2[categorical_cols] = df2[categorical_cols].astype('category')

# Verify the changes
print(df2.info())
```

```python
# Check for any dates that couldn't be parsed at all
print(f"Missing Order Dates: {df2['Order Date'].isna().sum()}")
print(f"Missing Ship Dates: {df2['Ship Date'].isna().sum()}")
```

```python
# Function with top X customers by sales filtered by multiple features (segment, city, category, sub_category, 
# ship_mode, region, start_date, end_date, month, year)

#If not imputed by user, return by default top 5

def top_customers(df2, top_x=5, segment=None, city=None, category=None, sub_category=None, ship_mode=None,
                    region=None, start_date=None, end_date=None, month=None, year=None):
    
    df_filtered = df2.copy()
    
    # Apply filters if provided
    if segment is not None:
        df_filtered = df_filtered[df_filtered["Segment"] == segment]
    if city is not None:
        df_filtered = df_filtered[df_filtered["City"] == city]
    if category is not None:
        df_filtered = df_filtered[df_filtered["Category"] == category]
    if sub_category is not None:
        df_filtered = df_filtered[df_filtered["Sub-Category"] == sub_category]
    if ship_mode is not None:
        df_filtered = df_filtered[df_filtered["Ship Mode"] == ship_mode]
    if region is not None:
        df_filtered = df_filtered[df_filtered["Region"] == region]
    if start_date is not None:
        df_filtered = df_filtered[df_filtered["Order Date"] >= pd.to_datetime(start_date)]
    if end_date is not None:
        df_filtered = df_filtered[df_filtered["Order Date"] <= pd.to_datetime(end_date)]
    if month is not None:
        df_filtered = df_filtered[df_filtered["Order Date"].dt.month == month]
    if year is not None:
        df_filtered = df_filtered[df_filtered["Order Date"].dt.year == year]

    # Top X customers with total sales
    top_customers = (df_filtered.groupby(["Customer ID", "Customer Name"])
                        ["Sales"].sum().reset_index().sort_values(by="Sales", ascending=False).head(top_x))

    return top_customers
```
