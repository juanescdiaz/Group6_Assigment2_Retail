Due to the missing values of City represent less than 4% of the total data, we will drop the rows with missing values in the City column.

```python
df_final = df.copy()
df_final['Customer Name'] = df_final['Customer Name'].fillna('Unknown')
```

```python
df_final.isna().sum()
```

```python
df2 = df_final.copy()

# Convert Dates
df2['Order Date'] = pd.to_datetime(df2['Order Date'], format= 'mixed')
df2['Ship Date'] = pd.to_datetime(df2['Ship Date'], format= 'mixed')


# Convert Financials (Cleaning strings like '$1,200.00' first)
df2['Sales'] = pd.to_numeric(df2['Sales'].replace(r'[\$,]', '', regex=True), errors='coerce')
df2['Profit'] = pd.to_numeric(df2['Profit'].replace(r'[\$,]', '', regex=True), errors='coerce')

# Convert Quantity to Integer
df2['Quantity'] = pd.to_numeric(df2['Quantity'], errors='coerce').fillna(0).astype(int)

# Clean up Postal Code (removing decimal .0 and treating as string)
df2['Postal Code'] = df2['Postal Code'].fillna(0).astype(int).astype(str)

# Optimize Categorical columns
categorical_cols = ['Ship Mode', 'Segment', 'Region', 'Category', 'Sub-Category']
df2[categorical_cols] = df2[categorical_cols].astype('category')

# Verify the changes
print(df2.info())
```

```python
# Check for any dates that couldn't be parsed at all
print(f"Missing Order Dates: {df2['Order Date'].isna().sum()}")
print(f"Missing Ship Dates: {df2['Ship Date'].isna().sum()}")
```

### Shipping Mode Usage Analysis

```python
#Copy the cleaning dataset to use in this specific analysis
df_s=df2.copy()
```

```python
#Task 24. Using the formula Ship Date - Order Date we can calculate the processing time of each Order
df_s['Processing Time']=(df_s['Ship Date']-df_s['Order Date']).dt.days
print(df_s.shape)
df_s.head()
```

```python
#Task 25. Counting the number of unique “Order Id” per shipping modes to identify the frequency and identify witch modes is more frequent to use.
order_counts = df_s.groupby('Ship Mode')['Order ID'].nunique().reset_index(name='Unique Order Count')
print(order_counts)
```

```python
#Task 26. Total of Unique Order Count per shipping mode divided by the Total of Unique Count. Percentage Value
total_orders = order_counts['Unique Order Count'].sum()
order_counts['Percentage'] = round((order_counts['Unique Order Count'] / total_orders) * 100, 2)
print(order_counts)
```

```python
#Task 27. sum the processing Time per shipping mode and then divides by the total of unique order count per shipping mode
# to get the average processing time per shipping mode
processing_time_sum = df_s.groupby('Ship Mode')['Processing Time'].sum().reset_index(name='Total Processing Time')
processing_time_sum = processing_time_sum.merge(order_counts[['Ship Mode', 'Unique Order Count']], on='Ship Mode')
processing_time_sum['Average Processing Time'] = round(processing_time_sum['Total Processing Time'] / processing_time_sum['Unique Order Count'], 2)
print(processing_time_sum[['Ship Mode', 'Average Processing Time']])
```

```python
#Task 28. Sum the total profit per shipping Mode
profit_sum = df_s.groupby('Ship Mode')['Profit'].sum().reset_index(name='Total Profit')
print(profit_sum)
```

```python
#Task 30. Merge all the calculated values in one unique Table for the report
report_table = order_counts.merge(processing_time_sum[['Ship Mode', 'Average Processing Time']], on='Ship Mode')
report_table = report_table.merge(profit_sum, on='Ship Mode')
report_table
```
