```python
df_final = df.copy()
df_final['Customer Name'] = df_final['Customer Name'].fillna('Unknown')
```

```python
df2 = df_final.copy()

# Convert Dates
df2['Order Date'] = pd.to_datetime(df2['Order Date'], format= 'mixed')
df2['Ship Date'] = pd.to_datetime(df2['Ship Date'], format= 'mixed')


# Convert Financials (Cleaning strings like '$1,200.00' first)
df2['Sales'] = pd.to_numeric(df2['Sales'].replace(r'[\$,]', '', regex=True), errors='coerce')
df2['Profit'] = pd.to_numeric(df2['Profit'].replace(r'[\$,]', '', regex=True), errors='coerce')

# Convert Quantity to Integer
df2['Quantity'] = pd.to_numeric(df2['Quantity'], errors='coerce').fillna(0).astype(int)

# Clean up Postal Code (removing decimal .0 and treating as string)
df2['Postal Code'] = df2['Postal Code'].fillna(0).astype(int).astype(str)

# Optimize Categorical columns
categorical_cols = ['Ship Mode', 'Segment', 'Region', 'Category', 'Sub-Category']
df2[categorical_cols] = df2[categorical_cols].astype('category')

# Clean 'Category' column by removing extra spaces and standardizing text
df2["Category"] = (
    df2["Category"]
    .astype(str)
    .str.replace(r"\s+", " ", regex=True)  
    .str.strip()                           
)

# Verify the changes
print(df2.info())
```

```python
# Function with top X customers by sales filtered by multiple features (segment, city, category, sub_category, 
# ship_mode, region, start_date, end_date, month, year)

#If not imputed by user, return by default top 5

def top_customers(df2, top_x=5, segment=None, city=None, category=None, sub_category=None, ship_mode=None,
                    region=None, start_date=None, end_date=None, month=None, year=None):
    
    df_filtered = df2.copy()
    
    # Apply filters if provided
    if segment is not None:
        df_filtered = df_filtered[df_filtered["Segment"] == segment]
    if city is not None:
        df_filtered = df_filtered[df_filtered["City"] == city]
    if category is not None:
        df_filtered = df_filtered[df_filtered["Category"] == category]
    if sub_category is not None:
        df_filtered = df_filtered[df_filtered["Sub-Category"] == sub_category]
    if ship_mode is not None:
        df_filtered = df_filtered[df_filtered["Ship Mode"] == ship_mode]
    if region is not None:
        df_filtered = df_filtered[df_filtered["Region"] == region]
    if start_date is not None:
        df_filtered = df_filtered[df_filtered["Order Date"] >= pd.to_datetime(start_date)]
    if end_date is not None:
        df_filtered = df_filtered[df_filtered["Order Date"] <= pd.to_datetime(end_date)]
    if month is not None:
        df_filtered = df_filtered[df_filtered["Order Date"].dt.month == month]
    if year is not None:
        df_filtered = df_filtered[df_filtered["Order Date"].dt.year == year]

    # Top X customers with total sales
    top_customers = (df_filtered.groupby(["Customer ID", "Customer Name"])
                        ["Sales"].sum().reset_index().sort_values(by="Sales", ascending=False).head(top_x))

    return top_customers
```

### Shipping Mode Usage Analysis

```python
#Copy the cleaning dataset to use in this specific analysis
df_s=df2.copy()
```

```python
#Task 24. Using the formula Ship Date - Order Date we can calculate the processing time of each Order
df_s['Processing Time']=(df_s['Ship Date']-df_s['Order Date']).dt.days
print(df_s.shape)
df_s.head()
```

```python
#Task 25. Counting the number of unique “Order Id” per shipping modes to identify the frequency and identify witch modes is more frequent to use.
order_counts = df_s.groupby('Ship Mode')['Order ID'].nunique().reset_index(name='Unique Order Count')
print(order_counts)
```

```python
#Task 26. Total of Unique Order Count per shipping mode divided by the Total of Unique Count. Percentage Value
total_orders = order_counts['Unique Order Count'].sum()
order_counts['Percentage'] = round((order_counts['Unique Order Count'] / total_orders) * 100, 2)
print(order_counts)
```

```python
#Task 27. sum the processing Time per shipping mode and then divides by the total of unique order count per shipping mode
# to get the average processing time per shipping mode
processing_time_sum = df_s.groupby('Ship Mode')['Processing Time'].sum().reset_index(name='Total Processing Time')
processing_time_sum = processing_time_sum.merge(order_counts[['Ship Mode', 'Unique Order Count']], on='Ship Mode')
processing_time_sum['Average Processing Time'] = round(processing_time_sum['Total Processing Time'] / processing_time_sum['Unique Order Count'], 2)
print(processing_time_sum[['Ship Mode', 'Average Processing Time']])
```

```python
#Task 28. Sum the total profit per shipping Mode
profit_sum = df_s.groupby('Ship Mode')['Profit'].sum().reset_index(name='Total Profit')
print(profit_sum)
```

```python
#Task 30. Merge all the calculated values in one unique Table for the report
report_table = order_counts.merge(processing_time_sum[['Ship Mode', 'Average Processing Time']], on='Ship Mode')
report_table = report_table.merge(profit_sum, on='Ship Mode')
report_table
```

```python
#bar chart for the percentage of unique orders per shipping mode
import matplotlib.pyplot as plt
bar_colors = ['tab:red', 'tab:purple', 'tab:blue', 'tab:pink']
plt.figure(figsize=(6, 4))
plt.bar(report_table['Ship Mode'], report_table['Percentage'], color=bar_colors)
plt.xlabel('Shipping Mode')
plt.ylabel('Percentage of Unique Orders')
plt.title('Distribution of Unique Orders by Shipping Mode')
plt.xticks(rotation=45)
plt.show()
```

```python
#bar chart for the average processing time per shipping mode
import matplotlib.pyplot as plt
bar_colors = ['tab:red', 'tab:purple', 'tab:blue', 'tab:pink']
plt.figure(figsize=(6, 6))
plt.bar(report_table['Ship Mode'], report_table['Average Processing Time'], color=bar_colors)
plt.xlabel('Shipping Mode')
plt.ylabel('Average Processing Time')
plt.title('Distribution of Average Processing Time by Shipping Mode')
plt.xticks(rotation=45)
plt.show()
```

```python
#bar chart for the total profit per shipping mode
import matplotlib.pyplot as plt
bar_colors = ['tab:red', 'tab:purple', 'tab:blue', 'tab:pink']
plt.figure(figsize=(6, 4))
plt.bar(report_table['Ship Mode'], report_table['Total Profit'], color=bar_colors)
plt.xlabel('Shipping Mode')
plt.ylabel('Total Profit')
plt.title('Distribution of Total Profit by Shipping Mode')
plt.xticks(rotation=45)
plt.show()
```
